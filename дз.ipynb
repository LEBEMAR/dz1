```python
import numpy as np
import matplotlib.pyplot as plt

alpha = np.linspace(0, 2 * np.pi, 1000)
r = 2 + np.sin(6 * alpha)
x = r * np.cos(alpha)
y = r * np.sin(alpha)


plt.figure(figsize=(6, 6))
plt.scatter(x,y, c=x, cmap = 'spring')
plt.plot(x, y, color='hotpink', linewidth=3)
plt.axis('equal')
plt.title("")
plt.grid(True, linestyle='dotted')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

amount = [50, 23, 24]
groups = ['group 1', 'group 2', 'group 3']
color_ = ['hotpink', 'violet', 'purple']

plt.pie(amount, labels = groups, colors = color_, autopct='%1.f%%')
plt.tittle('')
plt.show()
```

```python
from scipy.stats import norm

pos = 0
scale = 10
size = 1000
values = np.random.normal(pos, scale, size)
plt.hist(values, 100, density=True)
x = np.linspace(pos - 3*scale, pos + 3*scale, 100)
plt.plot(x, norm.pdf(x, pos, scale), 'r-', lw=2, label=' ')

plt.legend()
plt.show()
```
