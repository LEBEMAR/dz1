```python
import numpy as np
import matplotlib.pyplot as plt

alpha = np.linspace(0, 2 * np.pi, 1000)
r = 2 + np.sin(6 * alpha)
x = r * np.cos(alpha)
y = r * np.sin(alpha)


plt.figure(figsize=(6, 6))
plt.scatter(x,y, c=x, cmap = 'spring')
plt.plot(x, y, color='hotpink', linewidth=3)
plt.axis('equal')
plt.title("")
plt.grid(True, linestyle='dotted')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

amount = [50, 23, 24]
groups = ['group 1', 'group 2', 'group 3']
color_ = ['hotpink', 'violet', 'purple']

plt.pie(amount, labels = groups, colors = color_, autopct='%1.f%%')
plt.title('')
plt.show()
```

```python
import matplotlib.pyplot as plt

services = ['C', ' ', '', ' ', '']
satisfaction = [5.5, 1.5, 9.8, 4.0, 8.1]

plt.figure(figsize=(10, 7))
bars = plt.bar(services, satisfaction, color=['green', 'blue', 'purple', 'red', 'orange'],
               edgecolor='black', linewidth=1.2, alpha=0.9)

for bar, value in zip(bars, satisfaction):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1,
             f'{value}', ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.title('  ', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('', fontsize=12)
plt.ylabel('  ( 10)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.ylim(0, 10)

plt.tight_layout()
plt.show()
```
